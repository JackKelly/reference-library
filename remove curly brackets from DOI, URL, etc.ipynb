```python
from bibutils import load_bib, write_bib

BIB_FILENAME = 'mendeley-processed.bib'
#BIB_FILENAME = 'library.bib'

bib_database = load_bib(filename=BIB_FILENAME)
len(bib_database.entries)
```

```python
SUBSTRINGS_TO_REMOVE = ["{\\", "{", "}"]
```

```python
def remove_curly(string):
    for substring_to_remove in SUBSTRINGS_TO_REMOVE:
        string = string.replace(substring_to_remove, "")
    return string
```

```python
for entry in bib_database.entries:
    for key in ["url", "doi"]:
        try:
            value = entry[key]
        except KeyError:
            continue
        sanitised = remove_curly(value)
        if value != sanitised:
            print(entry['ID'], key, value, sanitised, "\n", sep="\n")
            entry[key] = sanitised
```

```python
write_bib(database=bib_database, filename=BIB_FILENAME)
```
